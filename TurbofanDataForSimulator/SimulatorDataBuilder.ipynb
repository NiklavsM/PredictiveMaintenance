```python
import pandas as pd
import numpy as np
import time as t
```

```python
time = int(t.time())
test_dataset = pd.read_csv('PredictiveMaintenanceEngineValidation.csv')
test_dataset = test_dataset.drop(columns=['label1', 'label2', 'cycle_norm', 'id', 'cycle', 's1','s5','s16','s18', 's19', 'setting3','s10', 'RUL'])
test_dataset = test_dataset.rename_axis('ID').values
pca_dataset = pd.read_csv('validationPCA.csv')
pca_dataset = pca_dataset.rename_axis('ID').values

# test_dataset.head(4)
```

```python
# Nor
out = ""
i = 0
id = 0
for row in test_dataset:
    time = time +60
    if i == 50:
        break
    out = out +"(\'" +str(time) + "\', \'" + str(id) + "\',"
    j = 0
    for item in row:
        if j == 16:
            out = out + "\'" + str(item) + "\'"
        else:
            out = out + "\'" + str(item) + "\',"
        j += 1
    id += 1
        
    out = out + "),\n"
    i += 1
print(out)
        
```

```python
# PCA Nasa Data
out = ""
i = 0
id = 0
for row in pca_dataset:
    time = time +60
    if i == 600:
        break
    out = out +"(\'" +str(time) + "\', \'" + str(id) + "\',"
    j = 0
    for item in row:
        if j == 10:
            out = out + "\'" + str(item) + "\'"
        else:
            out = out + "\'" + str(item) + "\',"
        j += 1
    id += 1
        
    out = out + "),\n"
    i += 1
print(out)
```
