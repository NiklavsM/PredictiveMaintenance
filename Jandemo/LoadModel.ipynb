```python
import tensorflow as tf
from tensorflow import keras
from keras.models import model_from_json
from keras import layers
from keras.models import Model
#print(tf.keras.__version__)
#print(tf.__version__)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math as mt
import random as rng

from sklearn import preprocessing
```

```python
json_file = open('model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("model.h5")
print("Loaded model from disk")
```

```python
valData = np.load("valData.npy")
valTarget = np.load("valTarget.npy")
```

```python
loaded_model.compile(optimizer=tf.keras.optimizers.RMSprop(0.001),
              loss='mse', 
              metrics=['mae', 'mse'])
```

```python
results = loaded_model.evaluate(valData,valTarget)
print("Model Mean Squared Error = " + str(results[0]))
print("Model Mean Absolute Error = " + str(results[1]))
```

```python
print("Running predictions...")
for i in range(100):
    test = np.array([valData[i]])
    prediction = loaded_model.predict(test, verbose=1)
    print("Predicted value = " + str(prediction) + ", Actual Value = " + str(valTarget[i]))
    print()
```
