```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
alarms = pd.read_csv("LDT_Alarm_Log.csv")
data = pd.read_csv("LDT-1Hz-Met-2017-02.csv")
alarms.head()
```

```python
data.head()
```

```python
alarms["TimeOn"] = pd.to_datetime(alarms["TimeOn"])
data["TimeStamp"] = pd.to_datetime(data["TimeStamp"])
alarms.index =  pd.to_datetime(alarms["TimeOn"])
data.index =  pd.to_datetime(data["TimeStamp"])
```

```python
tolerance = pd.Timedelta("1 second")
```

```python
alarms = alarms.sort_index()
```

```python
merged = pd.merge_asof(data, alarms, left_index=True, right_index=True, direction="nearest", tolerance=tolerance)
merged.sort_index()
```

```python
merged.keys
```

```python
merged.groupby("EventCode").head()
```

```python
errs = list(merged["EventCode"])
counter = 0
pointer = 0
time_til_err = []
for err in errs:
    if np.isnan(err):
        counter += 1
    else:  
        for i in range(counter, 0, -1):
            time_til_err.append(int(i))
        counter = 0
    pointer += 1

diff = pointer - len(time_til_err)
for i in range(0, diff):
    time_til_err.append(np.nan)
    

        
        
        
```

```python
merged.insert(1, "TimeUntilAlarm", time_til_err, True)
merged.head()
```

```python
plt.figure()
plt.title("General Cortime_til_errions for data")
corr = merged.corr()
sns.heatmap(corr)
plt.show()
```
